# Distribution Plots

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

* distplot
* jointplot
* pairplot
* rugplot
* kdeplot

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
plt.scatter(tips['total_bill'],tips['tip'])
plt.grid('True')
plt.ylabel('Tips')
plt.xlabel('Total bills')
plt.title('Bills Vs Tips');
```

# Distplot

```python
sns.distplot(tips['total_bill']);
```

```python
sns.distplot(tips['total_bill'],kde=True)
```

```python
sns.distplot(tips['total_bill'],kde=False);
```

# Histplot

```python
sns.set()
sns.histplot(tips['total_bill']);
```

```python
sns.histplot(tips['total_bill'],kde=True);
```

```python
sns.histplot(tips['total_bill'],kde=False);
```

## Jointplot

jointplot() allows you to basically match up two distplots for bivariate data. With your choice of what kind parameter to compare with:

  *  scatter
  *  reg
  *  resid
  *  kde
  *  hex



```python
sns.jointplot( x = 'total_bill', y = 'tip', data = tips);
```

```python
sns.jointplot( x = 'total_bill', y = 'tip', data = tips, kind = 'scatter');
```

```python
sns.jointplot(x = 'total_bill', y = 'tip',data = tips, kind = 'reg');
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid')
```

```python
sns.jointplot( x = 'total_bill', y = 'tip', data = tips, kind='kde');
```

```python
sns.jointplot( x = 'total_bill', y = 'tip', data = tips,kind='hex');
```

# Pairplot

```python
plt.figure(figsize=(15,15))
sns.pairplot(tips, hue = 'smoker', palette = 'husl');
```

```python
sns.pairplot(tips,hue='smoker',palette='rainbow');
```

```python
sns.pairplot(tips,hue='smoker',palette='coolwarm');
```

# Rugplot

```python
sns.rugplot(tips['total_bill']);
```

```python
sns.rugplot(tips['tip']);
```

# Kdeplots

```python
sns.set()
sns.kdeplot(tips['total_bill']);
```

```python
sns.kdeplot(tips['tip'])
```
