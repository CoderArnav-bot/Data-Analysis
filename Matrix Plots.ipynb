# Correlation

Correlation is a statistical technique that can show whether and how strongly pairs of variables are related. For example, height and weight are related; taller people tend to be heavier than shorter people

Correlation states how the features are related to each other or the target variable.

Correlation can be positive (increase in one value of feature increases the value of the target variable) or negative (increase in one value of feature decreases the value of the target variable).

Heatmap makes it easy to identify which features are most related to the target variable, we will plot the Heatmap of correlated features using the seaborn library.


```python
t.corr()
```

```python
f.corr()
```

```python
sns.heatmap(t.corr());
```

```python
sns.heatmap(f.corr())
```

```python
sns.heatmap(t.corr(),cmap='coolwarm',annot=True);
```

```python
sns.heatmap(t.corr(),cmap='rainbow',annot=True,fmt='0.2%');
```

```python
c = pd.read_csv('Car_sales.csv')
c.head()
```

```python
c['Latest Launch'] = pd.to_datetime(c['Latest Launch'])
```

```python
c.drop(columns=['Unnamed: 15','XX'],inplace=True)
```

```python
c.head()
```

```python
c['Year'] = c['Latest Launch'].dt.year
```

```python
c['Year']
```

```python
c_pivot = c.pivot_table(values='Sales in thousands',index = 'Manufacturer',columns = 'Year')
```

```python
sns.heatmap(c_pivot,cmap='rainbow',annot =True);
```

```python
u = pd.read_csv('Ufo.csv')
u.head()
```

```python
u['Time'] = pd.to_datetime(u['Time'])
```

```python
u['Year'] = u['Time'].dt.year
```

```python
u['Month'] = u['Time'].dt.month_name()
```

```python
u.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
u['Shape'] = le.fit_transform(u['Shape Reported'])
```

```python
u['Shape']
```

```python
u_pivot = u.pivot_table(values='Shape',index = 'Month',columns = 'Year')
```

```python
u_pivot
```

```python
sns.heatmap(u_pivot,cmap='Blues');
```
