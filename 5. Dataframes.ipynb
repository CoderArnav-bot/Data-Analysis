# Discovering and Deleting Duplicate Values

subset: Subset takes a column or list of column label. Its default value is none. After passing columns, it will consider them only for duplicates. keep: keep is to control how to consider duplicate value. It has only three distinct value and default is first.

   * If first, it considers first value as unique and rest of the same values as duplicate.
   * If last, it considers last value as unique and rest of the same values as duplicate.
   * If False, it consider all of the same values as duplicates



```python
data = pd.read_csv('employees.csv')
data
```

```python
data.shape
```

```python
data[data.duplicated(subset='First Name')]
```

```python
data.duplicated(subset='First Name')
```

```python
data.duplicated(subset='First Name')==True
```

```python
data.duplicated(subset='First Name',keep='first')
```

```python
data.duplicated(subset='First Name',keep='last')
```

```python
data.duplicated(subset='First Name',keep=False)
```

```python
data[data.duplicated(subset='First Name',keep='last')]
```

```python
data.drop_duplicates(subset='First Name',keep=False)
```

```python
data.drop_duplicates(subset=['First Name','Team'],keep='first')
```
