```python
car = pd.read_csv('Car_sales.csv')
car.head()
```

```python
car.drop(["Unnamed: 15","XX"],axis=1,inplace=True)
```

```python
car.columns
```

```python
car.info()
```

```python
for i in range(15):
    if car[car.columns[i]].dtype == 'float64':
        car[car.columns[i]].replace(np.nan,car[car.columns[i]].mean(),inplace=True)
```

```python
car.info()
```

```python
car['Manufacturer'].replace(np.nan,'BMW',inplace=True)
```

```python
car.info()
```

```python
car.head()
```

```python
car.groupby('Manufacturer')[['Price in thousands','Sales in thousands','Model']].head()
```

```python
car.groupby('Manufacturer').mean()
```

```python
car.groupby('Manufacturer').sum()
```

```python
car[car['Sales in thousands']==car['Sales in thousands'].min()][['Model','Price in thousands']].style.hide_index()
```

```python
car[car['Sales in thousands']==car['Sales in thousands'].max()][['Model','Price in thousands']].style.hide_index()
```

```python
car.nlargest(5,'Sales in thousands',keep='all')[['Price in thousands','Fuel efficiency']]
```

```python
car.nsmallest(7,'Sales in thousands',keep='all')[['Price in thousands','Vehicle type']]
```
