# Outliers

```python
sales = pd.read_csv('Sales_Outliers.csv')
```

```python
def detect_outliers(data):
    outlier = []
    th_value = 3
    mean = np.mean(data)
    std = np.std(data)
    for i in data:
        z_score = (i-mean)/std
        if np.abs(z_score) < th_value:
            outlier.append(i)
    return outlier
```

```python
y = detect_outliers(sales['Item_Outlet_Sales'])
y
```

```python
np.min(y)
```

```python
import matplotlib.pyplot as plt
plt.boxplot(sales['Item_Outlet_Sales'],showmeans = True)
plt.ylabel('Frequency')
plt.xlabel('Outliers');
```
