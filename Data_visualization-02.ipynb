# Creating Multiplots on Same Canvas

```python
# plt.subplot(nrows, ncols, plot_number)
#plt.tight_layout() can be used to free some spaces in-between the plots.
plt.subplot(1,5,1)
plt.tight_layout()
plt.plot(x,y,'c--');

plt.subplot(1,5,2)
plt.tight_layout()
plt.plot(y, x, 'g^-');

plt.subplot(1,5,3)
plt.tight_layout()
plt.plot(x,z,'y*-');

plt.subplot(1,5,4)
plt.tight_layout()
plt.plot(z,y,'r-');

plt.subplot(1,5,5)
plt.tight_layout()
plt.plot(x,y,'o--');
```

```python
fig = plt.figure()
plt.subplot(6,1,1)
plt.tight_layout()
plt.plot(x,y,'r--');

plt.subplot(6,1,2)
plt.tight_layout()
plt.plot(x,y,'b^-');


plt.subplot(6,1,3)
plt.tight_layout()
plt.plot(x,y,'gold');

plt.subplot(6,1,4)
plt.tight_layout()
plt.plot(x,y,'c-');

plt.subplot(6,1,5)
plt.tight_layout()
plt.plot(x,y,'orange');

fig.savefig('Multiple_plots.png')
```
